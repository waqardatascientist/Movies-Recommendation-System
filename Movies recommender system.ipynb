```python
import ast
```

```python
#this will extrect name of genres from the list of dictionary
#ast.literal_eval is used for the string operation
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['genres'][0]
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter+=1
        else:
            break
            
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast'][0]
```

```python
#now we extract the name of director from the crew column
# in the job dictionary it find the director and then name of director from here
def fetch_dir(dirr):
    L=[]
    for i in ast.literal_eval(dirr):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_dir)
```

```python
movies['crew'][1]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'][0] #we convert the list of string into seprate 
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
movies['overview']=movies['overview'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
#we cancatinate it these columnns into a single coulmln by the name of tags
movies['tags']=movies['cast']+movies['genres']+movies['overview']+movies['crew']
```

```python
new_movies_df = movies[['id', 'title', 'tags']]
```

```python
#now we convert the tags into the lower case
#new_movies_df['tags']=new_movies_df['tags'].apply(lambda x:x.lower())
new_movies_df['tags'] = new_movies_df['tags'].apply(lambda x: ' '.join(x).lower()if isinstance(x, list) else x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_movies_df['tags']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(new_movies_df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#this will provied us a list of movies that's similarity with other movies
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_movies_df[new_movies_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_movies_df.iloc[i[0]].title)
```
